```python
from docent._llm_util.providers.openai import get_azure_openai_client_async, get_openai_client_async

client = get_azure_openai_client_async()

async def minimal_chat_completion():
    response = await client.chat.completions.create(
        model="gpt-4o",  # or your Azure OpenAI deployment name
        messages=[
            {"role": "system", "content": "You are a helpful assistant."},
            {"role": "user", "content": "Hello, who are you?"}
        ],
        max_tokens=16,
    )
    print(response.choices[0].message.content)

await minimal_chat_completion()
```
