```python
from inspect_ai.log import read_eval_log, write_eval_log
log = read_eval_log("/Users/mengk/Code/luce-artifacts/mengk/inspect_logs/intercode-4o-default-scaffold.eval")
```

```python
import random

random.seed(1301)

new_samples = []
n_correct, n_incorrect = 0, 0
for s in log.samples:
    if s.epoch == 1:
        if s.scores["includes"].value != "C":
            if random.random() < 0.3:
                new_samples.append(s)
                n_incorrect += 1
        elif s.scores["includes"].value == "C":
            if random.random() < 0.075:
                new_samples.append(s)
                n_correct += 1


print(len(new_samples))
print(n_correct, n_incorrect)
log.samples = new_samples
```

```python
write_eval_log(log, "log.eval")
```
