```python
import json
with open("out_4o_4o-reasoning_gitignore.json", "r") as f:
    data = json.load(f)["data"]
```

```python
from docent.data_models import AgentRun

runs: dict[str, list[AgentRun]] = {}

for config_run in data:
    config = config_run["config"]
    runs[config["system_prompt"]] = [AgentRun(**run) for run in config_run["results"]]
```

```python
all_runs = []
for cur_runs in runs.values():
    all_runs.extend(cur_runs)
```

```python
all_sample_ids = list(set(run.metadata.get("sample_id") for run in all_runs))

import random
from uuid import uuid4
random.seed(1234)
ok_sample_ids = set(random.sample(all_sample_ids, 100))

sampled_runs = [run for run in all_runs if run.metadata.get("sample_id") in ok_sample_ids]
for run in sampled_runs:
    run.id = str(uuid4())
    for t in run.transcripts.values():
        t.id = str(uuid4())
```
